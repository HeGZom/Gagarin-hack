```python
from catboost import CatBoostClassifier, Pool
```

```python
dataset = pd.read_csv("preprocessed_data_1.csv")
dataset = dataset[dataset['sentiment_score'] != 0] 
train, test = train_test_split(dataset, test_size=0.2)
```

```python
train
```

```python
df_train = df.loc[train.index][['message_text', 'sentiment_score']]
df_test = df.loc[test.index][['message_text', 'sentiment_score']]
```

```python
y_train, X_train = df_train['sentiment_score'], df_train.drop(['sentiment_score'], axis=1)
```

```python
y_test, X_test = df_test['sentiment_score'], df_test.drop(['sentiment_score'],axis=1)
```

```python
train_pool = Pool(
    data=X_train,
    label=y_train,
    text_features=['message_text']
)
test_pool = Pool(
    data=df_test[[c for c in df_test.columns if c!='sentiment_score']][['message_text']], 
    label=df_test['sentiment_score'],
    text_features=['message_text']
)
```

```python
torch.cuda.is_available
```

```python
from sklearn.metrics import accuracy_score
```

```python
from hyperopt import hp, fmin, tpe, Trials, STATUS_OK
from hyperopt.pyll.base import scope

def objective(params):
    model = CatBoostClassifier(**params, task_type='GPU')
    
    model.fit(train_pool, eval_set=(test_pool), verbose=100)
    
    y_pred = model.predict(test_pool)
    accuracy = accuracy_score(y_test, y_pred)
    
    return {'loss': -accuracy, 'status': STATUS_OK}

# Пространство поиска всех гиперпараметров
space = {
    'depth': hp.quniform('depth', 1, 6, 1),
    'learning_rate': hp.loguniform('learning_rate', -3, 0),
    'l2_leaf_reg': hp.loguniform('l2_leaf_reg', -3, 3),
    'iterations': hp.quniform('iterations', 100, 5000, 100),
    'border_count': hp.quniform('border_count', 1, 255, 1),
    'thread_count': hp.quniform('thread_count', 1, 16, 1),
}


# Запуск оптимизации гиперпараметров
trials = Trials()
best = fmin(objective, space, algo=tpe.suggest, max_evals=20, trials=trials)

print(best)
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased',do_lower_case=False)
```

```python
df1=df.drop_duplicates()

# Переиндексация DataFrame после удаления дубликатов
df1=df.reset_index(drop=True)
len(df1)
```

```python
segments_ids_ans = list()
indexed_tokens_ans = list()
for i in tqdm(range(len(df1['message_text']))):
    text = df1['message_text'][i]
    marked_text = "[CLS]" + text + "[SEP]"
    tokenized_text = tokenizer.tokenize(marked_text)
    segments_ids = [1] * len(tokenized_text)
    indexed_tokens = tokenizer.convert_tokens_to_ids(tokenized_text)
    segments_ids_ans.append(segments_ids)
    indexed_tokens_ans.append(indexed_tokens)
```

```python
sentences_count = len(df1['message_text'])
tokens_tensor_ans = [torch.tensor([indexed_tokens_ans[i]]) for i in tqdm(range(sentences_count))]
segments_tensors_ans = [torch.tensor([segments_ids_ans[i]]) for i in tqdm(range(sentences_count))]
```

```python
encoded_layers_ans = list()
with torch.no_grad():
    for i in tqdm(range(sentences_count)):
        try:
            encoded_layers, _ = model(tokens_tensor_ans[i], segments_tensors_ans[i])
            encoded_layers_ans.append(encoded_layers)
        except IndexError:
            print("Ошибка IndexError на индексе:", i)
```
